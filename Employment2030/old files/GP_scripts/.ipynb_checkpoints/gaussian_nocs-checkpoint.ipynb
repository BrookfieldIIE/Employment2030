```python
#read full noc set
file = "../tables/noc_scores.csv"
nocs = pd.read_csv(file,index_col=['noc'])
nocs.sort_index(inplace=True)
nocs.drop(['Unnamed: 0'],axis=1,inplace=True)
disc_nocs = np.round(nocs).astype(int)
cont_nocs = nocs
```

```python
features = ['value.Monitoring',
            'value.Social Perceptiveness',
            'value.Service Orientation',
            'value.Systems Evaluation',
            'value.Inductive Reasoning',
            'value.Memorization',
            'value.Flexibility of Closure',
            'value.Perceptual Speed',
            'value.Finger Dexterity',
            'value.Multilimb Coordination',
            'value.Near Vision',
            'value.Far Vision',
            'value.Computers and Electronics',
            'value.Chemistry',
            'value.Transportation']
work_agg = ['work_num']
work_ind = ['workshop.number']
```

```python
mappings = {0: 'increase',
            1: 'constant/not increase',
            2: 'decrease'}
```

```python
def train_full_data(x_all, y_all, kern, name, agg_level):
    print('Training full model:', name)
    start = time.time()
    if agg_level == 'ind': all_features = features + work_ind
    else:                  all_features = features + work_agg
    x, y = x_and_y(x_all[all_features], y_all)
    x = normalize(x)
    kernel = get_kern(kern len(x[0]))
    likelihood = create_likelihood(y_all)
    model = build_model(x, y, likelihood, kernel)
    print('Model trained')
    predictions, classes = new_nocs(model, disc_nocs[features], name)
    end = time.time()
    print('Time elapsed (sec):', end - start)
    return model, predictions, classes
```

```python
def new_nocs(model, nocs, name):
    predictions = {}
    classes = defaultdict(int)
    for index, row in nocs.iterrows():
        workshop_preds = [0, 0, 0]
        for i in range(1, 7):
            prediction = [x for [x] in predictive_density(model, np.asarray(row + [i]))]
            prediction = scale_pred(prediction)
            workshop_preds[np.argmax(prediction)] += 1
        category = mappings[np.argmax(workshop_preds)]
        predictions[index] = {'Category':     category,
                              'Distribution': prediction}
        classes[category] += 1
    print('Saving:', name)
    pickle_results(predictions, name)
    return predictions, classes
```
